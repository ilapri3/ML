## 2. Создайте функцию, которая возвращает диагональную матрицу, по диагонали которой располагаются члены арифметической прогрессии. Пусть эта функция принимает три значения: размер матрицы, первый элемент и разность арифметической прогрессии. Выведите матрицу с конкретными значениями параметров. Напечатайте эту матрицу, используя print.

```python
def matrix(size, first, diff):
    diag_el = [first + i*diff for i in range(size)]
    diag_matrix = np.diag(diag_el)
    return diag_matrix
matrix(5, 1, 3)
```

```python
A = matrix(5, 1, 3)
print(A)
```

## 4. Создайте функцию, которая возвращает квадратную матрицу, содержащую единицы на границе и нули внутри. Пусть эта функция принимает размер матрицы.

```python
def quadrat(size):
    a = np.zeros(size)
    a[:, 0] = 1
    a[0, :] = 1
    a[:, -1] = 1
    a[-1, :] = 1
    return a

quadrat([5,5])
```

## 5. Создайте две матрицы размером 5 $\times$ 5. Первая матрица должна содержать цифру 5 в шахматном порядке, остальные значения - нули. Вторая матрица - треугольная, состоящая из цифры 5 на местах с ненулевыми значениями. Посчитайте их детерминанты и найдите обратные матрицы (если это возможно).

```python
matrix1 = np.zeros((5, 5))
matrix1[0::2, 0::2] = 5
matrix1[1::2, 1::2] = 5
matrix1
```

```python
matrix2 = np.zeros((5, 5))
for i in range(5):
    for j in range(i + 1):
        matrix2[j, i] = 5
matrix2
```

```python
np.linalg.det(matrix1)
```

```python
np.linalg.det(matrix2)
```

```python
try:
    inv_matrix1 = np.linalg.inv(matrix1)
    print('/nОбратная матрица')
    print(inv_matrix1)
except np.linalg.LinAlgError:
    print('Не существует обратной матрицы')
```

```python
np.linalg.inv(matrix2)
```

## 17. Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.

```python
labels = first_400_raws['Rooms'].value_counts().index # вариации с квартирами

sizes = first_400_raws['Rooms'].value_counts().values #количество вариантов для каждой вариации

most = [0.05 if size == max(sizes) else 0 for size in sizes] #массив для отделения сектора

plt.figure(figsize = (10, 5))
plt.pie(sizes, explode = most, labels = labels, autopct='%1.1f%%')
plt.title('Количество квартир в зависимости от количества комнат')
plt.show()
```

## 18. Постройте гистограмму по целевой переменной Price. Оцените визуально, по какой цене продаётся наибольшее количество квартир.

## 19. Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

```python
useful_data = first_400_raws[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']]

sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize = (12, 10))

sns.scatterplot(x = 'Rooms', y = 'Price', data = useful_data, ax = axes[0, 0])
axes[0, 0].set_title('Rooms vs Price')

sns.scatterplot(x = 'Square', y = 'Price', data = useful_data, ax = axes[0, 1])
axes[0, 1].set_title('Square vs Price')

sns.scatterplot(x = 'HouseFloor', y = 'Price', data = useful_data, ax = axes[1, 0])
axes[1, 0].set_title('HouseFloor vs Price')

sns.scatterplot(x = 'HouseYear', y = 'Price', data = useful_data, ax = axes[1, 1])
axes[1, 1].set_title('HouseYear vs Price')

plt.tight_layout()
plt.show()
```

## 22. При помощи сетки графиков PairGrid визуализируйте попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния.
```python
df = first_400_raws[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']]
graph = sns.PairGrid(df)

graph.map_diag(sns.histplot)
graph.map_lower(sns.kdeplot)
graph.map_upper(sns.scatterplot)

plt.show()
```

## 23. Постройте тепловую карту матрицы корреляции (df.corr()) признаков Rooms, Square, HouseFloor, HouseYear, Price. По ней определите, какие признаки являются зависимыми (у таких признаков коэффициент корреляции близок к единице).

```python
matrix = first_400_raws[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']]
cor_matrix = matrix.corr()

plt.figure(figsize=(10, 5))
sns.heatmap(cor_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Тепловая карта матрицы корреляции')
plt.show();
```
